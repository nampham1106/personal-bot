## Business Categories

```python
# load businesses
business_df = pd.read_parquet(f"{DATASET_DP}/yelp_academic_dataset_business_sample_100_biz.parquet")
```

```python
business_df
```

```python
categories = []
for ls in business_df['categories'].str.split(', '):
    categories.extend(ls)
```

```python
categories
```

```python
from collections import Counter 
(
    pd.DataFrame.from_dict(dict(Counter(categories)), orient='index', columns=['count'])
    .assign(
        total=business_df.shape[0],
        perc=lambda df: df['count']/df['total']
    )
    .sort_values(['count'], ascending=False)
    .style
    .format("{:,.1%}", subset=['perc'])
    .bar(subset=['count', 'perc'])
)
```
